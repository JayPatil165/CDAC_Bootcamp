```python
import pandas as pd
```

```python
sales_dt = pd.read_csv('Sales.csv')
```

```python
sales_dt.head(7)
```

```python
sales_dt['Revenue'] = sales_dt['Price'] * sales_dt['QuantitySold'] # Calculate the revenue column.
```

```python
sales_dt.head(7)
```

```python
category_sorted = pd.get_dummies(sales_dt, columns=['ProductCategory']) # Sort by category.
```

```python
category_sorted.head(7)
```

```python
Q1 = category_sorted['Price'].quantile(0.25)
Q3 = category_sorted['Price'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Use IQR to detect outliers
```

```python
cleaned_sales = category_sorted[(category_sorted['Price'] >= lower_bound) & category_sorted['Price'] <= upper_bound] # Remove outliers.
```

```python
cleaned_sales.head(7)
```

```python
from sklearn.preprocessing import MinMaxScaler # For min max scaling the Price and Revenue Columns.
```

```python
cleaned_sales[['Price', 'Revenue']] = MinMaxScaler().fit_transform(cleaned_sales[['Price', 'Revenue']])
```
