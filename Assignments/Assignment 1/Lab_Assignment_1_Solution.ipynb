```python
import pandas as pd
```

```python
student_dt = pd.read_csv('Students.csv')
```

```python
student_dt.head()
```

```python
student_dt['MathScore'].fillna(student_dt['MathScore'].mean(), inplace = True)
#Expty ones will he filled with the value provided. Here the average of the whole column.
#inplace = True will actually change the value of the table and not give it a copy to change it.
```

```python
student_dt['EnglishScore'].fillna(student_dt['EnglishScore'].mean(), inplace = True)
# THe mean value of the whole column is filled into the empty ones. Same as done above.
```

```python
student_dt.head() # We can see empty cells / entries filled.
```

```python
from sklearn.preprocessing import LabelEncoder
# Library needed to encode the Female and Male in the Gender column with 0 and 1.
```

```python
student_dt.head() # Used head as it only has 5 values.
```

```python
student_dt['Gender'] = LabelEncoder().fit_transform(student_dt['Gender']) # THe LableEncoder() function encodes the given data. As female comes before male, the encoder tkaes 0 for female by default. But to change it we need to specify.
```

```python
student_dt.head() # Here the Gender Encoding can be seen.
```

```python
student_dt['TotalScore'] = student_dt['MathScore'] + student_dt['EnglishScore'] 
```

```python
student_dt.head() # Corrected values for TotalScore are seen below.
```

```python
from sklearn.preprocessing import StandardScaler # Will be used to scale the values. That is standardize them.
```

```python
student_dt[['MathScore', 'EnglishScore', 'TotalScore']] = StandardScaler().fit_transform(student_dt[['MathScore', 'EnglishScore', 'TotalScore']]) # standardizes the data.
```
