```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression  # linearRegression model is used from sklearn.
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score  # These matrices are used as we have regression matrix.
```

```python
california_housing = fetch_california_housing()
```

```python
california_housing  # Gives info of the data.
```

```python
#Create data frame
data = pd.DataFrame(data=np.c_[california_housing['data'], california_housing['target']], columns=california_housing['feature_names'] + ['target'])
data # prints out the data
```

```python
x = data.drop("target", axis=1)
y = data["target"]
```

```python
x = data[["MedInc", "HouseAge"]]
y = data["target"]
```

```python
x
```

```python
#Splitting the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Create LiearRegression Model
model = LinearRegression()
```

```python
#Fit data into model / train the model 
model.fit(x_train, y_train)
```

```python
# Make predictions 
y_pred = model.predict(x_test)
```

```python
# Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)
```

```python
# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
```

```python
# R2 Score
r2 = r2_score(y_test, y_pred)
```

```python
# Print Results
print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R-squared (RÂ²) Score: {r2:.2f}')
```

```python
# Visualization
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_test, color='red', alpha=0.7, label='Actual House Prices')
plt.scatter(y_test, y_pred, color='orange', alpha=0.7, label='Predicted House Prices')
plt.xlabel('Actual House Prices')
plt.ylabel('Predicted House Prices')
plt.title('Actual vs. Predicted House Prices')
plt.grid(True)
plt.legend()
plt.show()
```
