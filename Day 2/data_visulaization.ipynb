```python
import matplotlib.pyplot as plt # Library for data visualization
import pandas as pd
import numpy as np
```

```python
x = np.array([1,3,4,5])
y = np.array([6,7,8,9])
```

```python
print(x,y)
```

```python
#----- Line Chart -----#
plt.plot(x,y) # The plot looks weird as the arrays are randomly generated.
plt.xlabel('X-Axis', weight='bold')
plt.ylabel('Y-Axis', weight='bold')
plt.title("Line Chart", weight='bold')
```

```python
#------ Bar Graph -----#
Languages = ["Python", "C", "C++", "Java", "Flutter"]
No_of_Employees = [ 65, 70, 60, 82, 15]
bargraph = plt.bar(Languages, No_of_Employees, color='g', edgecolor = 'b', width=0.1, linewidth=1,         align='center', ) # Width changes the width of the plotted bar graph.
plt.xlabel("Number of Employees", weight='bold')
plt.ylabel("Language", weight='bold')
plt.title("Bar Graph", weight='bold')
bargraph[0].set_color('c')
bargraph[1].set_color('r')
bargraph[2].set_color('g')
bargraph[3].set_color('b')
bargraph[4].set_color('y')
# Instead of coloring them like this we could use 'color=['c', 'r', 'g', 'b', 'y']' in the plot.bar() statement.
plt.legend(bargraph, No_of_Employees)  # Here legend takes the plot first and data later.
plt.show()
```

```python
#------ Scatter Plot ------#
# Sample data
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]
colors = [1,3,4,5,6,7,8,9,10,11]
sizes = [100,150,200,250,300,350,400,450,500,550]

plt.scatter(x, y, c=colors,s=sizes,cmap='plasma', marker='o')

plt.xlabel('X-axis', weight='bold')
plt.ylabel('Y-axis', weight='bold')
plt.title('Scatter Plot', weight='bold')
plt.colorbar()
plt.show()
```

```python
#------ Pie Chatrt -----#
a =  [15, 25, 20, 15, 20]
b = ['C', 'C++', 'Python', 'Java', 'Flutter']

plt.pie(a, labels = b,explode=[0,0,0,0.1,0], labeldistance=1.1, autopct='%1.1f%%', radius=1.1) # explode made the java thing pop out. autopct shows percentage values.
plt.title("Pie Chat", weight='bold')
plt.show()
```

```python
#----- Histogram -----#
histogram_data = np.random.randn(1000) # Generate random data for histogram
plt.hist(histogram_data, bins=100,color='c', edgecolor='black', align='mid') #bins divides data into these many sets\blocks.

plt.xlabel('X-Axis', weight='bold')
plt.ylabel('Y-Axis', weight='bold')
plt.title('Histogram', weight='bold')

plt.show()
```
