```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_from_excel = pd.read_excel('supermarket_sales.xlsx')
data_from_excel.shape
data_from_excel.info()
data_from_excel.describe()
```

```python
data_from_excel.head()
```

```python
#----- Count Plot -----#
plt.figure(figsize=(15,5))
sns.countplot(x='Product line',data=data_from_excel) # The data from the Product line comlumn gets made into a count plot.
plt.xlabel('X-Axis', weight='bold')
plt.ylabel('Y-Axis', weight='bold')
plt.title("Count Map", weight='bold')
plt.show()
```

```python
#----- Horizontal Count Plot -----#
sns.countplot(y="Product line", data = data_from_excel) # Plots the data horizontally.
plt.xlabel('X-Axis', weight='bold')
plt.ylabel('Y-Axis', weight='bold')
plt.title("Horizontal Count Map", weight='bold')
plt.show()
```

```python
#----- Count Plot with hue sorting -----#
plt.figure(figsize=(15,5))
sns.countplot(x="Product line", hue="Gender", data = data_from_excel, palette ='deep') # hue splits the data as per the specified column.
plt.xlabel('X-Axis', weight='bold')
plt.ylabel('Y-Axis', weight='bold')
plt.title("Splitted Count Map", weight='bold')
plt.show()
```

```python
# Use color palette to change the color of the plot.
plt.figure(figsize=(15,5))
sns.countplot(x="Product line", data = data_from_excel,
facecolor = (0,1,1,0.3), # Changes the color of the plot.
linewidth=3,
edgecolor= 'black',
palette= 'pastel')
plt.legend()
plt.xlabel('X-Axis', weight='bold')
plt.ylabel('Y-Axis', weight='bold')
plt.title("Count Map with changed Color", weight='bold')
plt.show()
```

```python
#------ HeatMap ------#
data_for_heatmap = np.random.rand(10,15)
plt.figure(figsize=(10, 6))  # Set the figure size
sns.heatmap(data_for_heatmap, cmap='viridis', annot=True)  # cmap sets the color of map (google or more color options), annot adds the values.
plt.title('Heatmap', weight='bold')
plt.xlabel('Columns', weight='bold')
plt.ylabel('Rows', weight='bold')
# plt.colorbar() # Not required as it already does give a color bar.
plt.show()
```

```python
#----- BoxPlot -----#
data_for_boxplot = np.random.rand(10,15)
plt.figure(figsize=(15,5))
sns.boxplot(data=data_for_boxplot, width=0.7, palette='Blues')
plt.title('Box Plot', weight='bold')
plt.xlabel('X-Axis', weight='bold')
plt.ylabel('Y-Axis', weight='bold')
plt.show()
```
